```python
data_path = '../../data/p300'
data_file = 'p300_wiko_06-08-25'
```

```python
data = UnicornData(f'{data_path}/{data_file}.xdf', delay=0.073)
epochs = data.filter().create_epochs(event_dict=dict(standard=1, oddball=2), tmin=-0.2, tmax=1.5)
```
